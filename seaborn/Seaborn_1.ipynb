```python
# kdplot->
# its also a axises function graph

tips=sns.load_dataset("tips")
tips
sns.kdeplot(data=tips,x="total_bill")
# density is the probablity 
```

```python
# displot ->kdeplot
sns.displot(data=tips,x="total_bill",kind="kde")
```

```python
# hue
# sns.displot(data=tips,x="total_bill",kind="kde",col="sex",)
sns.displot(data=tips,x="total_bill",kind="kde",hue="sex")
```

```python
# rugplot
# this is not so usefull 
sns.kdeplot(data=tips,x="total_bill")
sns.rugplot(data=tips,x="total_bill")

# rugplot is also part of probablity or density
```

```python
# Bivariate -Histogram
sns.histplot(data=tips,x="total_bill",y="tip")
sns.displot(data=tips,x="total_bill",y="tip",kind="hist",hue="day")

plt.grid()
```

```python
# Bivariate Kdeplot
sns.kdeplot(data=tips,x="total_bill",y="tip")
plt.grid()
plt.show()
```

```python
# matrix graph
gap=px.data.gapminder()
gap
```

```python
life=gap.pivot(index="country",columns="year",values="lifeExp")
```

```python
# axeslevel function
plt.figure(figsize=(15,15))
sns.heatmap(data=life)
plt.show()
```

```python
live_3=gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")

plt.figure(figsize=(15,15))
sns.heatmap(data=live_3,annot=True,linewidths=0.5,cmap="summer")
plt.show()
```
