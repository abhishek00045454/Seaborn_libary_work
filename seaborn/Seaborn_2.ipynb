```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
iris=sns.load_dataset("iris")
```

```python
# Categorical Plots
```

```python
# strip plot
# axes level function 
sns.stripplot(data=tips,x="day",y="total_bill",jitter=False)

# its type of scatter plot
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="strip")
# figure level plot
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="strip",jitter=0.2,hue="sex")
```

```python
# swarm plot

sns.catplot(data=tips,x="day",y="total_bill",kind="swarm")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))

sns.swarmplot(data=tips,x="day",y="total_bill")
```

```python
plt.figure(figsize=(10,10))
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm",col="sex",hue="sex")
```

```python
# Ditrubtion graph
# Box plot
# its imp in ml
```

```python
sns.boxplot(data=tips,x="day",y="total_bill")
sns.boxplot(data=tips,x="sex",y="total_bill")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="box",hue="sex")
```

```python
# single boxplot ->numerical plot
sns.boxplot(data=tips,y="total_bill",hue="sex")
```

```python
# vloin(kdeplot+boxplot)
sns.violinplot(data=tips,x="day",y="total_bill")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="violin",hue="sex",spilt=True)
```

```python
# barplot
import numpy as np
# sns.barplot(data=tips,x="smoker",y="total_bill",hue="sex",estimator=np.median)
sns.barplot(data=tips,x="smoker",y="total_bill",hue="sex",estimator=np.min)
```

```python
# point  bar
sns.pointplot(data=tips,x="sex",y="total_bill",hue="smoker",estimator=np.min,errorbar=None)
```

```python
# countplot
```

```python
sns.countplot(data=tips,x="sex",hue="day")
```

```python
# facteing plot
sns.catplot(data=tips,x="sex",y="total_bill",col="smoker",kind="box")
```

```python
# its also axes level
# hue plot is not passed
sns.regplot(data=tips,x="total_bill",y="tip")
sns.regplot(data=tips,x="total_bill",y="tip",hue="sex")
```

```python
# lmplot
sns.lmplot(data=tips,x="total_bill",y="tip",hue="sex")
```

```python
# residual plot -> this plot tell us linear regression that can be calculate wrong output that tell us residual plot
sns.residplot(data=tips,x="total_bill",y="tip")
```

```python
# FaceGrid
# figure plot-> relplot->displot->catplot->lmplot

sns.catplot(data=tips,x="sex",y="total_bill",kind="box",col="day",row="time")
```

```python
# Facegrid

g=sns.FacetGrid(data=tips,col="day",row="time",hue="smoker")
g.map(sns.boxplot,"sex","total_bill")
g.add_legend()
```

```python
# pair plot
sns.pairplot(data=iris,hue="species",diag_kind="hist")
```

```python
# pair grid
g=sns.PairGrid(data=iris,hue="species")
g.map(sns.scatterplot)
```

```python
sns.pairplot(data=iris,hue="species")
g.map_diag(sns.violinplot)
g.map_offdiag(sns.histplot)
```

```python
plt.figure(figsize=(10,10))
g=sns.PairGrid(data=iris,hue="species")
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
plt.figure(figsize=(10,10))
g=sns.PairGrid(data=iris,hue="species",vars=["sepal_width","petal_width"])
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
# joint polt and joint grid
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip",kind="hist",hue="sex")
```

```python
g=sns.JointGrid(data=tips,x="total_bill",y="tip",hue="sex")
g.plot(sns.scatterplot,sns.histplot)
```
